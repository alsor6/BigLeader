```python
import math
import os
import copy
import numpy as np
from tkinter import *
from tkinter import messagebox

## 함수
import sys 
sys.setrecursionlimit(262144)

def displayImage() :
    global window, canvas, paper, height, width, filename, median
    if canvas != None :
        canvas.destroy()

    canvas = Canvas(window, height=512, width=512)
    paper = PhotoImage(height=512, width=512)
    canvas.create_image((512 / 2, 512 / 2), image=paper, state='normal')

    for i in range(height) :
        for k in range(width) :
            r = g = b = image[i][k]
            paper.put('#%02x%02x%02x' % (r, g, b), (k, i))

    canvas.pack()
    
#1.반전
def reverseImage() :
    for i in range(height):
        for k in range(width):
            image[i][k] = 256 - image[i][k]

    displayImage()
    
#2.밝게
def lightImage():
    for i in range(height):
        for k in range(width): 
            if image[i][k] + 50 > 256 :
                image[i][k] = 256
            else:
                image[i][k] += 50
    displayImage()
    
#3.어둡게
def darkImage():
    for i in range(height):
        for k in range(width): 
            if image[i][k] - 50 < 0 :
                image[i][k] = 0
            else:
                image[i][k] -= 50
    displayImage()
    
#4.흑백(127)
def BlackandWhiteImage():
    for i in range(height):
        for k in range(width):
            if image[i][k] < 127:
                image[i][k] = 0
            else:
                image[i][k] = 256
    displayImage()

#5.흑백(중앙값)
#5-1 1차원 배열로 바꾸기
def imageto1array(image):
    array = []
    for i in range(height):
        for k in range(width):
            array.append(image[i][k])
    return array

#5-2 퀵 정렬
def quick_sort2(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    equal = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort2(left) + equal + quick_sort2(right)

    
#5-3 중앙값 찾기
def findmedian(image):
    image2 = imageto1array(image)
    #quick_sort(image2, 0, len(image2)-1)
    image3 = quick_sort2(image2)
    length = len(image3)
    if length % 2 == 1:
        median = image3[length // 2]
    else:
        mid1 = image3[length // 2]
        mid2 = image3[(length // 2) - 1]
        median = (mid1 + mid2) / 2
    return median


#5-4 중앙값 이용해서 흑백변환
def BlackandWhiteImage2():
    median = findmedian(image)
    for i in range(height):
        for k in range(width):
            if image[i][k] < median : 
                image[i][k] = 0
            else:
                image[i][k] = 255             
    displayImage()

##회전    
#-----------------------90도---------------------------------            
def rotateImage90():
    global height, width, image

    # 이미지 크기 변경
    new_height = width
    new_width = height
    new_image = [[0 for _ in range(new_width)] for _ in range(new_height)]

    # 이미지 회전
    for i in range(height):
        for j in range(width):
            new_image[j][height - i - 1] = image[i][j]
    image = new_image
    displayImage()

#-----------------------180도---------------------------------
    
def rotateImage180():
    global height, width, image

    # 이미지 회전
    
    new_image = [[0 for _ in range(width)] for _ in range(height)]
    for i in range(height):
        for j in range(width):
            new_image[height-i-1][width-j-1] = image[i][j]
    image = new_image
    displayImage()
                     
#--------------270도------------------------------------------
def rotateImage270():
    global height, width, image
    
    # 이미지 크기 변경
    new_height = width
    new_width = height
    new_image = [[0 for _ in range(new_width)] for _ in range(new_height)]
    
    # 이미지 회전
    for i in range(height):
        for j in range(width):
            new_image[width-j-1][i] = image[i][j]
    image = new_image
    displayImage()
    
#------------------------------------------------------------------------

##미러링
def leftrightImage() :
    image2 = copy.deepcopy(image)
    for i in range(height):
        for k in range(width):
            if k < (width / 2):
                image[i][k] = image[i][width - k - 1]
            else:
                image[i][k] = image2[i][width - k - 1]
    displayImage()
    
def updownImage() :
    image2 = copy.deepcopy(image)
    for i in range(height) :
        for k in range(width) :
            if i < (height/2) :
                image[i][k] = image[height-i-1][k]
            else :
                image[i][k] = image2[height-i-1][k]
    displayImage()


    
##감마연산, 파라볼라 연산
def gammaImage() :
    global image
    image2 = copy.deepcopy(image)
    gamma = 1.5
    for i in range(height):
        for k in range(width):
            image[i][k] = int(np.power(image2[i][k] / 255.0, gamma) * 255)
    displayImage()

def paraImage() : #cup파라볼라
    global image
    image2 = copy.deepcopy(image)
    for i in range(height):
        for k in range(width):
            image[i][k] = int(255 * np.power(image2[i][k] / 127.0, 2))
    displayImage()


## 변수
window, canvas, paper = None, None, None
filename = ""
height, width = 0, 0
image = []

## 메인
window = Tk()
window.geometry('600x800')
window.title('영상처리 Alpha')

btn1 = Button(window, text ='반전이다', command=reverseImage)
btn1.pack()
btn2 = Button(window, text ='밝아져라', command=lightImage)
btn2.pack()
btn3 = Button(window, text ='어두워져라', command=darkImage)
btn3.pack()
btn4 = Button(window, text ='흑백이다', command=BlackandWhiteImage)
btn4.pack()
btn5 = Button(window, text ='중앙으로 만든 흑백이다', command=BlackandWhiteImage2)
btn5.pack()
btn6 = Button(window, text ='90도 돌아라', command=rotateImage90)
btn6.pack()
btn7 = Button(window, text ='180도 돌아라', command=rotateImage180)
btn7.pack()
btn8 = Button(window, text ='270도 돌아라', command=rotateImage270)
btn8.pack()
btn9 = Button(window, text ='좌우미러링', command=leftrightImage)
btn9.pack()
btn10 = Button(window, text ='상하미러링', command=updownImage)
btn10.pack()
btn11 = Button(window, text ='감마연산', command=gammaImage)
btn11.pack()
btn12 = Button(window, text ='파라볼라연산', command=paraImage)
btn12.pack()




filename = 'Etc_Raw(squre)/LENNA512.raw'
# 파일 크기 알아내기
fSize = os.path.getsize(filename) # Byte 단위
height = width = int(math.sqrt(fSize))
# 메모리 확보 (영상 크기)
image = [ [0 for _ in range(width)] for _ in range(height)]
# 파일 --> 메모리 로딩
rfp = open(filename, 'rb')
for i in range(height) :
    for k in range(width) :
        image[i][k] = ord(rfp.read(1))
        
# image2 = imageto1array(image)
# quick_sort(image2, 0, len(image2)-1)
# print(image2)
# qs = quick_sort2(image2)
# print(qs)



rfp.close()

displayImage()


window.mainloop()
```
